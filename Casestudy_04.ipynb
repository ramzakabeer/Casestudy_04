# Check whether all the zones generate the same amount of sales.

Here we've to compare more than two means, therefore we use a one-way ANOVA test.

```python
#Null hypothesis: All the zones generate the same amount of sales;H0: μ1=μ2=μ3=μ4
#Alternative hypothesis: At least, one zone sale is different;H1: All μ are not equal
```

```python
#calculating F-statistic and p-value
fvalue, pvalue = stats.f_oneway(data['Zone - A'], data['Zone - B'], data['Zone - C'], data['Zone - D'])
```

```python
print(fvalue, pvalue)
```

```python
#Degrees of Freedom
```

No:of groups, k = 4
DoF between = k-1 = 4-1 = 3
Total no:of samples in all groups,n = 116
DoF within = n-k = 116-4 = 112

```python
#calculating F-crtic
f_critic=stats.f.ppf(1-0.05,3,112)
f_critic
```

```python
#Hypothesis Testing
```

```python
if(fvalue>f_critic and pvalue<0.05):
    print('We reject the null hypothesis. That is, zone-wise sales are not same.')
else:
    print('We failto reject the null hypothesis. That is, all the zones generate the same amount of sales.')
```
